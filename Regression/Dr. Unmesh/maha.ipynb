```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
# from google.colab import files
import matplotlib.pyplot as plt
```

```python
# Upload the CSV file
data_all = pd.read_csv("Maharashtra.csv")
```

```python
data_all.head()
```

```python
data = data_all[(data_all['Age'] > 17) & (data_all['Age'] < 61)]
data.head()
```

```python
# Define the target variable (y) and features (X)
# Replace 'target_column' with the name of your target column
target_column = "Household'S Usual Consumer Expenditure In A Month (Rs.)"
X = data.drop(columns=[target_column])
y = data[target_column]
```

```python
X.head()
```

```python
y.head()
```

```python
# Create and train the regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Make predictions
y_pred = model.predict(X)
```

```python
# Evaluate the model
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
# Plot actual vs predicted values
plt.scatter(y, y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.show()
```

```python
import statsmodels.api as sm

# Add a constant to the features (for the intercept term)
X_with_const = sm.add_constant(X)

# Fit the regression model
model_sm = sm.OLS(y, X_with_const).fit()

# Display the summary of the regression analysis
print(model_sm.summary())
```
