**COMPARISON OF IDFC & HDFC BANK OVER 5 YEARS**

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Define stock symbols for HDFC and IDFC First Bank
hdfc_symbol = "HDFCBANK.NS"
idfc_symbol = "IDFCFIRSTB.NS"

# Define time period
start_date = "2020-01-01"
end_date = "2025-01-01"

# Fetch historical data
df_hdfc = yf.download(hdfc_symbol, start=start_date, end=end_date)
df_idfc = yf.download(idfc_symbol, start=start_date, end=end_date)
```

```python
df_idfc.head()
```

```python
df_hdfc.head()
```

```python
# Plot stock prices
plt.figure(figsize=(12, 6))
plt.plot(df_hdfc.index, df_hdfc['Close'], label='HDFC Bank', color='steelblue')
plt.plot(df_idfc.index, df_idfc['Close'], label='IDFC First Bank', color='red')
plt.xlabel('Date')
plt.ylabel('Stock Price (INR)')
plt.title('Stock Price Comparison: HDFC vs IDFC First Bank')
plt.legend()
plt.grid()
plt.show()
```

```python
# Calculate daily returns
df_hdfc['Returns'] = df_hdfc['Close'].pct_change()
df_idfc['Returns'] = df_idfc['Close'].pct_change()

# Plot returns
plt.figure(figsize=(12, 6))
plt.plot(df_hdfc.index, df_hdfc['Returns'], label='HDFC Bank Returns', color='blue', alpha=0.6)
plt.plot(df_idfc.index, df_idfc['Returns'], label='IDFC First Bank Returns', color='red', alpha=0.6)
plt.xlabel('Date')
plt.ylabel('Daily Returns')
plt.title('Stock Returns Comparison: HDFC vs IDFC First Bank')
plt.legend()
plt.grid()
plt.show()
```

The above visualizations, clearly answer why when comparing two stocks, returns are considered as opposed to stock prices.
* Returns normalize performance regardless of the starting price.
* Returns measure the percentage change in value, making them comparable across different stocks.
* A $100 stock increasing to $120 (+20%) performs the same as a $10 stock rising to $12 (+20%), even though their prices differ significantly.
* Example
  
    Stock A: Starts at $50, ends at $60  Return = +20%

    Stock B: Starts at $200, ends at $220  Return = +10%
    
    Even though Stock B increased by $20 in price, Stock A actually had a higher return (20% vs. 10%), making it a better-performing stock in relative terms.
